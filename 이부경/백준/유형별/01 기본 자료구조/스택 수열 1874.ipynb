```python
# 직접 풀었다는 거 말고는 의미 없는 코드입니다.
# 처음에 접근을 잘못해서 복잡해졌어요.

N = int(input())
n = N
num_stack = []
s = []
i = 0

def get_input():
    global n
    if n > 0:
        n -= 1
        return int(input())        
    return 0

goal_num = get_input()
while i < N:
    while True:
        s.append('+')
        i += 1
        num_stack.append(i)
        if num_stack[-1] == goal_num:
            break
        if i > N:
            break

    while len(num_stack) > 0 and num_stack[-1] == goal_num:
        s.append('-')
        num_stack.pop()        
        goal_num = get_input()
        
while True:
    if get_input() == 0:
        break

if len(num_stack) != 0:
    print('NO')
else:
    print('\n'.join(s))
```

```python
# 강의 들었던 것 안 보고 다시 짜봤습니다.
def solve():
    s = []
    result = []
    i = 1
    for _ in range(int(input())):
        x = int(input())

        while i <= x:
            result.append('+')
            s.append(i)
            i += 1

        if s[-1] == x:
            result.append('-')
            s.pop()
        else:
            return 'NO'
        
    return '\n'.join(result)

result = solve()
print(result)
```
