```python
tree = {}

class Tree:
    def __init__(self, name, left, right):
        self.name = name
        self.left = left
        self.right = right

def PreorderTraversal(name):
    if name == '.':
        return

    t = tree[name]    
    print(t.name, end='')
    PreorderTraversal(t.left)
    PreorderTraversal(t.right)

def InorderTraversal(name):
    if name == '.':
        return
    
    t = tree[name]
    InorderTraversal(t.left)
    print(t.name, end='')
    InorderTraversal(t.right)

def PostorderTraversal(name):
    if name == '.':
        return

    t = tree[name]
    PostorderTraversal(t.left)
    PostorderTraversal(t.right)
    print(t.name, end='')

for _ in range(int(input())):
    t = Tree(*(input().split()))
    tree[t.name] = t

PreorderTraversal('A')
print()
InorderTraversal('A')
print()
PostorderTraversal('A')
print()
```
